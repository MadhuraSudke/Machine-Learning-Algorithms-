Normalizing the datset 

```python
def mean(X):
    return np.sum(X)/len(X) # find the mean
```

```python
# find the variance 
def variance(X):
    mean_value = mean(X)
    return np.sum((X- mean_value)**2)/len(X)
```

```python
# find the norm 
def norm(X):
    mean_value = mean(X)
    variance_value = variance(X)
    return (X-mean_value)/np.sqrt(variance_value)
```

```python
X_norm = norm(X)
```

```python
X_norm
```

```python
plt.plot(X_norm, Y)
```

Reshape the data

```python
X_norm = X_norm.reshape(-1,1) 
"""Reshaping 2D array with one column, -1 mean figure of the no of rowns automatically  m,n m= no of data point , n= no of features"""
```

```python
reg = LinearRegression().fit(X_norm,Y)
""" Creating a training model where x_norm is input data and y is target 
slope(M): how much model Y change when X change 
intercept(b) Y when X =0"""
```

```python
reg.score(X_norm,Y) 
"""Return R square (coefficient of determination)"""
```

```python
reg.coef_
"""Return slope(m) of the regression line."""
```

```python
Y_pred = reg.predict(X_norm)
# train that model to predict the Y values for given X values
```
